```python
#modeling
X = df[["Average Cost for two", "Has Table booking", "Has Online delivery", "Price range"]]
```

```python
#Above 2 below 2.5 Poor
#Above 2.5 below 3.5 Average
#Above 3.5 below 4.0 Good
#Above 4.5 Excellent
```

```python
Y = df[["Aggregate rating"]]
```

```python
X
```

```python
X ["Price range"].value_counts ()
```

```python
#changing has table booking column to numerical values
from sklearn import preprocessing
labelencode = preprocessing.LabelEncoder()
```

```python
X["Has Table booking"] = labelencode.fit_transform(X["Has Table booking"])
```

```python
X
```

```python
labelencode.classes_
```

```python
labelencode.classes_[1]
```

```python
#Has Table booking 0 is No and 1 is Yes
```

```python
X["Has Online delivery"] = labelencode.fit_transform(X["Has Online delivery"])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size= 0.2)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np 
```

```python
def modelresults(predictions):
    print("Mean absolute error on model is {}".format(mean_absolute_error(y_test, predictions)))
    print("Root mean squared error model is {}".format(np.sqrt(mean_squared_error(y_test, predictions))))
```

```python
from sklearn.linear_model import LinearRegression 
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
predictionsfromlr = lr.predict(x_test)  
```

```python
modelresults(predictionsfromlr)
```

```python
predictionsfromlr
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV 
```

```python
svrmodel = SVR()
```

```python
param_gridsvr = {"C": [0.1,0.5], "kernel" : ["linear","poly"],"degree" : [2]}
```

```python
gridsvr = GridSearchCV(svrmodel, param_gridsvr)
gridsvr.fit(x_train,y_train)
```

```python
gridsvr.predict(x_test)
```

```python
predsgridsvr = gridsvr.predict(x_test)
```

```python
modelresults(predsgridsvr)
```

```python
param_grid = {
    "max_depth" : [10,30],
    "min_samples_leaf" : [1,2],
    "min_samples_split" : [2,5]
}
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
treemodel = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator= treemodel, param_grid=param_grid)
```

```python
grid_tree.fit(x_train, y_train)
```

```python
treepredictions = grid_tree.predict(x_test)
```

```python
modelresults(treepredictions)
```
